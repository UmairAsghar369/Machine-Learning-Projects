```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('study_exam_score_clean.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['hours_studied'])
y = DataFrame(data, columns=['exam_score'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X ,y)
plt.title('Studentâ€™s exam score based on the number of hours they studied')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.ylim(51,90)
plt.xlim(0,11)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X ,y)
```

```python
regression.coef_ # Slpoe
```

```python
regression.intercept_ # Intercept
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X ,y)
plt.plot(X,regression.predict(X),color='orange',linewidth=3)

plt.title('Studentâ€™s exam score based on the number of hours they studied')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.ylim(52,90)
plt.xlim(1,11)
plt.show()
```

```python
regression.score(X, y)
```
