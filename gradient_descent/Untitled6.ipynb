# Notebook import and packages

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Example 1 :A simple cost function
## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x +1
```

```python
# Make data
x_1 = np.linspace(start = -3, stop = 3, num = 500)
```

```python
# make plot

plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

# Slope and Derivatives

```python
def df(x):
    return 2*x + 1
```

```python
# Plot Function and derivative side by side
plt.figure(figsize = [12,5])
# 1 chart : Cost Function
plt.subplot(1,2,1)

plt.xlim([-3,3])
plt.ylim([0,8])
plt.title('Cost Function',fontsize = 17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1) ,color = "blue",linewidth = 3)

# 2 Chart : Derivative Function
plt.subplot(1,2,2) 
plt.grid()
plt.title('Slope of Cost Function' ,fontsize = 17)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)

plt.xlim(-2,3)
plt.ylim(-3,6)
plt.plot(x_1,df(x_1) , color = 'skyblue', linewidth = 4)

plt.show()
```

## Python loops and Gradient Desent 

```python
# for loop in python 
for n in range(5):
    print('Hello World ',n)
print("ENd of loop")
```

```python
 # python While Loop 
counter = 0 
while counter  < 5:
    print("Counting ...",counter)
    counter =  counter +1
print('Ready or not , Here i am :)')
```

```python
# Gradient
new_x = 3
previous_x = 0
step_multiplier = 0.1

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

print('Local minimum occures at :', new_x)
print('Slope or df(x) value at this point is :',df(new_x))
print('f(x)Value of cost at this point is :', f(new_x))
```

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001 

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    step_size = abs(new_x - previous_x)
    # print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if step_size < precision:
        print('our loop ran this many times ',n)
        break

print('Local minimum occures at :', new_x)
print('Slope or df(x) value at this point is :',df(new_x))
print('f(x)Value of cost at this point is :', f(new_x))
```

```python
 # Superimpose the gradient descent calculations on Plot
plt.figure(figsize = [15,5])
# 1 chart : Cost Function
plt.subplot(1,3,1)

plt.xlim([-3,3])
plt.ylim([0,8])
plt.title('Cost Function',fontsize = 17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1) ,color = "blue",linewidth = 3 , alpha = 0.8)

value = np.array(x_list)
plt.scatter(x_list,f(value) , color = 'red' , s = 100 ,alpha = 0.6)

# 2 Chart : Derivative Function
plt.subplot(1,3,2) 
plt.grid()
plt.title('Slope of Cost Function' ,fontsize = 17)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)

plt.xlim(-2,3)
plt.ylim(-3,6)
plt.plot(x_1,df(x_1) , color = 'skyblue', linewidth = 4 ,alpha = 0.6)

plt.scatter(x_list,slope_list ,color = 'red', s = 100 ,alpha = 0.5 )

# 3 Chart : Derivative Function (close up)
plt.subplot(1,3,3)

plt.grid()
plt.title('Gradient Descent (Close up)' ,fontsize = 17)
plt.xlabel('X',fontsize=16)

plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.plot(x_1,df(x_1) , color = 'skyblue', linewidth = 6 ,alpha = 0.6)

plt.scatter(x_list,slope_list ,color = 'red', s = 300 ,alpha = 0.6 )

plt.show()
```
