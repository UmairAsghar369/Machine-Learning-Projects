```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
```

# Scatter plot with styling

```python
plt.figure(figsize=(10,6))   
sns.scatterplot( data=iris,
    x ="sepal_length", 
    y ="petal_length", 
    hue ="species" , 
    s = 80, edgecolor = 'black'
)

plt.title("Sepal Length vs Petal Length (Iris Dataset)")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

# Adding Regression Line

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data = iris,
    x ="sepal_length",
    y ="petal_length",
    hue ="species",
    s = 80,                
    edgecolor = "black")

sns.regplot(
    data = iris,
    x ="sepal_length",
    y ="petal_length",
    scatter = False,  
    color = "red")

plt.title("Sepal Length vs Petal Length (Iris Dataset)", fontsize=14)
plt.xlabel("Sepal Length ")
plt.ylabel("Petal Length ")
plt.show()
```
