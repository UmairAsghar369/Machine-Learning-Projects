```python
import matplotlib.pyplot as plt
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [30, 32, 31, 29, 28, 27, 29]

plt.plot(days, temperature,  color='blue')
plt.title("Temperature Over 7 Days")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.show()
```

```python
categories = ["Food", "Transport", "Shopping", "Bills", "Others"]
expenses = [1500, 700, 1200, 800, 400]

plt.bar(categories, expenses, color ='green')
plt.title("Weekly Expenses")
plt.xlabel("Category")
plt.ylabel("Amount ")
plt.show()
```
