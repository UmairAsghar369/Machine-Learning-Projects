```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('9.1 lsd_math_score_data.csv')
```

```python
print(data)
```

```python
time = data['Time_Delay_in_Minutes']
LSD = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
type(LSD)
```

```python
type(time)
```

```python
%matplotlib inline

plt.plot(time,LSD, color = '#16a085',linewidth = 4)
plt.title('Tissue concentration over LSD', fontsize = 17)
plt.xlabel('Times in minutes',fontsize = 14)
plt.ylabel('Tissue LSD ppm',fontsize=14)
plt.text(x=0, y=-0.25, s='Wagner et al. (1968)',fontsize = 12)

# plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.ylim(1,7)
plt.xlim(0,500)

plt.style.use('classic')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
regr.fit(LSD,score)
```

```python
print('Theta1 :',regr.coef_[0][0])
print('Intersept :',regr.intercept_[0])
print('R-square :',regr.score(LSD,score))
predicted_score = regr.predict(LSD)
```

```python
plt.title('Arithmetic vs LSD-25', fontsize=17)
plt.xlabel('Tissue LSD ppm', fontsize=14)
plt.ylabel('Performance Score', fontsize=14)
plt.ylim(25, 85)
plt.xlim(1, 6.5)
plt.style.use('fivethirtyeight')
plt.scatter(LSD, score, color='blue', s=100, alpha=0.7)
plt.plot(LSD, predicted_score, color='red', linewidth=3)
plt.show()
```
